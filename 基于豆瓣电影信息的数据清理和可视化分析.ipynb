# 二、加载数据

```python
df_1 = pd.read_csv("./csv/top250_f1.csv",
                   sep="#")######问题1，分隔符是#，所以要设置#是分隔符
df_1
```

### 2.1、根据数据的分析，下哦那个两个数据集中提取信息

```python

df1_cut = df_1[['num','title','init_year','area','genre','rating_num','comment_num']]
df2_cut = df_2[['num','language','director','cast','movie_duration','tags']]
```

```python
df1_cut.head()###查看前五列数据
```

```python
####组合两个切出的数据，在num位置用outer方式连接
```

```python
df = pd.merge(df1_cut,df2_cut,how="outer",on='num')  #赋值给df
df
```

```python
#####################二、、清理数据格式
###################2.1清理genre，language，director，cast，movie_duration
```

```python
# deal_list=['genre','language','director','cast','movie_duration','tags']
# for i in deal_list:
#     df[i]=df[i].str[2:-2]



df['genre']=df.genre.str[2:-2]  #取出genre这一列的括号和引号，用切片方式
df['language'] = df['language'].str[2:-2]#同上
df['director'] = df['director'].str[2:-2]
df['cast'] = df['cast'].str[2:-2]
df['movie_duration'] = df['movie_duration'].str[2:-2]
df.head()
```

```python
######################2.2根据不同国家的电影数
####2.2.1区域内容清洗
```

```python
# 中国大陆 香港’ “中国大陆”和“香港”之间用空格隔开，可以用 str.split 
# 函数进行分列， 再应用apply(pd.Series)函数使其作用到每一行和列。
a = df["area"].str.split(" ").apply(pd.Series)
a.head()
```

```python
# 汇总统计，同时NaN值用0填充（缺失值处理）
area = a.apply(pd.value_counts).fillna(0) 
area.head()
```

```python
######将每个国家1到5列出现次数的总和相加
a = area.apply(lambda col :col.sum(),axis=1)
a.head()
```

```python
#包装成一个标准的DataFrame
area2 = pd.DataFrame(a, columns = ['counts'])
area2.sort_values(by="counts",ascending=False).head()#降序排列
```

#### 2.3对电影类型的处理

```python
# 对电影类型进行拆分（空格分隔）利用apply(pd.Series)应用到每一行
genre_split = df['genre'].str.split(' ').apply(pd.Series)
genre_split.head()
```

```python
# # 对每一列的值重合的作了一个汇总统计
genre=genre_split.apply(pd.value_counts).fillna(0)
genre.head()
```

```python
#求和
genre=genre.apply(lambda row :row.sum(),axis=1)
genre
```

```python
#转换为DataFrame
g=DataFrame(genre,columns=["counts"])
```

```python
#画图#电影类型的分类
g1=g.sort_values(by="counts",ascending=False)
g1.plot(kind="bar",figsize = (20,12),grid=True)
plt.show()
```

```python
g2 = g.unstack().dropna().reset_index()      #####行索引转换，变为层次索引，去掉NA值，最后转换Series变为DataFeame,并增加索引
g2
```

```python
from pyecharts import Bar
```

```python
############################2.4画图，电影类型
mybar=Bar("豆瓣电影剧情分类",width=2000,height=600)
attr=g.index
value=g.counts
mybar.add("",attr,value,mark_point=["max","min"],is_random=True)
mybar.render("豆瓣电影剧情分类.html")
mybar
```

#### 2.6电影时长

```python
movie_time = df["movie_duration"].str.strip().str.replace('\', \'','#').str.split('#').apply(pd.Series)
movie_time.head()###去空格，替换
```

```python
#有些电影时长存在多种版本，一般情况下第一个时长为国内最普通、观看数量较多的版本，因此仅取第一个时长。
duration = movie_time[0].str.split('分').apply(pd.Series)[0].str.strip()
duration
```

```python
# 检查异常值
duration[duration.str.len() > 3]
```

```python
# 观察数据发现有特殊的：Japan: 75 
#,为第244条。类似于这样的数据，处理掉
duration[244] = duration[244].split(' ')[1]
```

```python
#将分钟转化为整型
#将整理的数据重新赋值给duration
duration = duration.astype(int)
duration.dtypes
```

```python
df['movie_duration'] = duration
df['movie_duration'].head()
```

# 四、可视化分析

### 4.1、通过散点图查看电影排名与评分的关系

```python
plt.scatter(df['rating_num'], df['num']) #绘制散点图
plt.xlabel('rating_num')    #x轴标签
plt.ylabel('ranking list')  #y轴标签
plt.show()
```

```python
plt.figure(figsize=(14,6)) 
plt.subplot(1,2,1)
plt.scatter(df['rating_num'], df['num'])
plt.xlabel('rating_num')
plt.ylabel('ranking list')
plt.gca().invert_yaxis()  #获取当前的画板，并倒置y轴的坐标

plt.subplot(1,2,2)
plt.hist(df['rating_num'],bins = 15)  #划分多少个区间
plt.xlabel('rating_num')
plt.show()
```

### 4.2、评论人数和电影排名的关系

```python
plt.figure(figsize=(14,6)) 
plt.subplot(1,2,1)
plt.scatter(df['comment_num'], df['num'])
plt.xlabel('comment_num')
plt.ylabel('ranking list')
plt.gca().invert_yaxis() #道理同上

plt.subplot(1,2,2)
plt.hist(df['comment_num'])
plt.xlabel('comment_num')
plt.show()
```

```python
# 利用“泊松分布” 显示相关性
df['num'].corr(df['comment_num'])
```

##### 结论： 评价人数呈右偏分布，随评价人数的增多，豆瓣Top250排名名次有提前趋势，Pearson相关系数为 -0.6623 ，呈强相关性

### 4.3、排名与电影时长的关系

```python
plt.figure(1)
plt.figure(figsize=(14,6)) 
plt.subplot(1,2,1)
plt.scatter(df['movie_duration'], df['num'])
plt.xlabel('movie_duration')
plt.ylabel('ranking list')
plt.gca().invert_yaxis()

plt.subplot(1,2,2)
plt.hist(df['movie_duration'],bins = 50)
plt.xlabel('movie_duration')
plt.show()
```

```python
# 利用“泊松分布” 显示相关性
df['num'].corr(df['movie_duration'])
```
