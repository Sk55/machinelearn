### 第一步：导入相应的模块

```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']= ['SimHei'] #中文注释
plt.rcParams['axes.unicode_minus'] = False #显示正负号
```

### 第二步：利用pandas读取datasets目录下chipo.csv并显示前十行数据（赋值给变量chipo）

```python
chipo = pd.read_csv("./datasets/chipo.csv" )
chipo
```

### 第三步：根据列名为item_name中每种商品出现的频率，绘制出如图下图所示的柱状图
##### (注：方法多种多样，选择一种最适合你的即可)

```python
# # 对每一列的值重合的作了一个汇总统计
chipo["number"] = "1"

tu1=chipo["quantity"].groupby(chipo["item_name"]).sum()
tu2=DataFrame(tu1)
tu2
####排序的，哈哈哈哈哈哈哈哈哈哈哈哈哈
    
g1=tu2.sort_values(by="quantity",ascending=False)
g1.plot(kind="bar",figsize = (20,12))
plt.title('购买次数最多的商品排名')
plt.xlabel('商品名称')
plt.ylabel('出现的订单次数')
plt.legend(loc='upper right')
```

```python
chipo["number"] = "1"

tu1=chipo["quantity"].groupby(chipo["item_name"]).sum()
tu2=DataFrame(tu1)
tu2
####不排序的，哈哈哈哈哈哈哈哈哈哈哈哈哈
tu3=tu2.head()
g1=tu3.sort_values(by="quantity",ascending=False)
g1.plot(kind="bar")
plt.title('购买次数最多的商品排名')
plt.xlabel('商品名称')
plt.ylabel('出现的订单次数')
plt.legend(loc='upper right')
```

### 第四步：根据订单编号（odrder_id）进行分组,求出每个订单花费的总金额，例如订单编号为1的总金额为11.56美元。然后根据每笔订单的总金额和每笔订单购买商品的总数量画出散点图（总金额为x轴，商品总数为y轴），结果图如下所示：

```python
money = lambda x: float(x[1:])  #去掉符号
chipo['price'] = chipo['item_price'].agg(money)
data=DataFrame([chipo["order_id"],chipo["price"]])
data=data.T
ss=data.groupby('order_id')['price'].sum()
cc=data.groupby('order_id').count()
x=ss.values
y=cc.values
# plt.figure(figsize=(15,7))
plt.scatter(x,y,label='商品名',color='r',s=50,marker='o')
plt.title('每笔订单总金额和商品订单关系')
plt.xlabel('总金额')
plt.ylabel('商品总数')
# plt.legend(loc='upper right')
```
