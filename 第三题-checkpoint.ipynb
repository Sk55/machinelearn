### 第一步：导入相关模块

```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
import scipy as sp
from matplotlib import pyplot as plt
from pyecharts import Bar
from pyecharts import WordCloud

import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"]=['SimHei']  # 用于正常显示中文标签
plt.rcParams['axes.unicode_minus']=False  # 用来正常显示负号
```

### 第二步：加载datasets下的tips.csv文件数据，并显示前五行记录

```python
df = pd.read_csv('datasets/tips.csv')
df.head(5)
```

### 第三步：绘制消费金额频率分布直方图

```python
#绘制直方图
plt.hist(df['total_bill'],bins=15)
plt.xlabel('频率')
plt.ylabel('消费总金额')
plt.title('消费金额频率分布直方图')
```

### 第四步：绘制总消费金额与消费金额的散点图关系

```python
real_bill = df['total_bill']-df['tip']
plt.scatter(df['total_bill'],real_bill)
plt.xlabel('消费总金额')
plt.ylabel('消费金额')
plt.title('消费金额和消费总金额的散点图')
```

### 第五步：在同一图中绘制出吸烟顾客与不吸烟顾客的消费金额与小费之间的散点图关系

```python
smoker=df[df['smoker'] !='No']

# 不吸烟顾客
no_smoker = df[df['smoker'] =='No']

#绘制散点图
# 创建一行两列的画板
plt.figure(figsize=(14,6))
#选中一行两列的第一个区域
plt.subplot(1,2,1)
plt.scatter(smoker['total_bill'],smoker['tip'])
plt.xlabel('消费总金额')
plt.ylabel('小费金额')
plt.title('吸烟顾客')

#选中第二个区域
plt.subplot(1,2,2)
plt.scatter(no_smoker['total_bill'],no_smoker['tip'])
plt.xlabel('消费总金额')
plt.ylabel('小费金额')
plt.title('不吸烟顾客')
```

### 第六步：在同一图中绘制出女性与男性中吸烟与不吸烟顾客的消费金额与小费之间的散点图关系

```python
#吸烟
s=df[df['smoker'] !='No']
man = s[smoker['sex']=='Male']
Female = s[s['sex']=='Female']

#不吸烟
no = df[df['smoker'] =='No']
man2= no[no['sex']=='Male']
Female2 = no[no['sex']=='Female']

#新建
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.scatter(man['total_bill'],man['tip'],label='男性吸烟顾客')
plt.scatter(man2['total_bill'],man2['tip'],label='男性不吸烟顾客')
plt.xlabel('消费总金额')
plt.ylabel('小费金额')
plt.title('sex=Male')
plt.legend()

#2
plt.subplot(1,2,2)
plt.scatter(Female['total_bill'],Female['tip'],label='女性吸烟顾客')
plt.scatter(Female2['total_bill'],Female2['tip'],label='女性不吸烟顾客')
plt.xlabel('消费总金额')
plt.ylabel('小费金额')
plt.title('sex=Female')
plt.legend()
```
